# Application of Matplotlib 

A simple example

```python
import pandas as pd
data = pd.read_excel('data.xlsx').iloc[:,0:2]
data.head()
```

```python
data = data.set_index('SPTR Index')
#data.index = pd.to_datetime(data.index)
data.columns=['price']
data.head()
```

# 1. 修改图像属性

```python
# 坐标
# 更改坐标轴范围
a = [148, 147, 132, 138, 150, 122, 93, 109]
plt.plot(a)
#plt.xlim()
plt.ylim(140,150)
plt.show()
```

```python
# 设定坐标标签与显示角度
plt.plot(a)
plt.ylim(140,150)
plt.xticks(range(8),['2015-02-01','2015-02-02','2015-02-03','2015-02-04',
                     '2015-02-05','2015-02-06','2015-02-07','2015-02-08']
          ,rotation = 45)
plt.yticks()
plt.show()
```

```python
# 添加文本
# 添加标题
plt.plot(a)
plt.xticks(range(8),['2015-02-01','2015-02-02','2015-02-03','2015-02-04',
                     '2015-02-05','2015-02-06','2015-02-07','2015-02-08'],
          rotation = 45)
#plt.title('My score')
plt.title('My score',loc='right')
plt.show()
```

```python
# 设定坐标轴标签
plt.plot(a)
plt.title('My score',loc='center')
#plt.title('My score',loc='left')
plt.xlabel('date')
plt.ylabel('score')
plt.xticks(range(8),['2015-02-01','2015-02-02','2015-02-03','2015-02-04',
                     '2015-02-05','2015-02-06','2015-02-07','2015-02-08'],
          rotation = 45)
plt.show()
```

```python
# 增加图形背景
plt.plot(a)
plt.title('My score',loc='center')
plt.xlabel('date')
plt.ylabel('score')
plt.xticks(range(8),['2015-02-01','2015-02-02','2015-02-03','2015-02-04',
                     '2015-02-05','2015-02-06','2015-02-07','2015-02-08'],
          rotation = 45)
plt.grid(True)
plt.show()
```

```python
# 增加图例
b = [96, 98, 78, 89, 67, 67, 53, 44]
c = [100]*8
plt.plot(a,label='maths')
plt.plot(b,label='physics')
plt.plot(c,label='english')
plt.title('My score',loc='center')
plt.xlabel('date')
plt.ylabel('score')
plt.xticks(range(8),['2015-02-01','2015-02-02','2015-02-03','2015-02-04',
                     '2015-02-05','2015-02-06','2015-02-07','2015-02-08'],
          rotation = 45)
#plt.legend()
#plt.legend('best')
plt.legend(loc='best')
plt.show()
```

```python
# 多种线条属性 linestyle / ls
plt.plot(a,label='maths',ls='-')
plt.plot(b,label='physics',ls='-.')
plt.legend()
plt.show()
```

```python
# color/c 'r' 'b' 'g' 'y' 'k'
plt.plot(a,label='maths',ls='-',color='red')
plt.plot(b,label='physics',ls='-.',c='g')
plt.legend()
plt.show()
```

```python
plt.plot(a,label='maths',marker='s',c='r')
plt.plot(b,label='physics',marker='v',c='black')
plt.legend()
plt.show()
```

```python
plt.plot(a,label='maths',lw = 2)
plt.plot(b,label='physics', linewidth=10)
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = plt.subplot(211) #这里的函数名称和之前的plt不一样的
ax1.set_title('Graph 1')
ax1.set_xlabel('Count')
ax1.plot(a)

ax2 = plt.subplot(212)
ax2.set_title('Graph 2')
ax2.set_ylabel('Score')
ax2.bar(range(8),a,color='red')
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = plt.subplot(211)
ax2 = plt.subplot(212)
ax1.set_title('Graph 1')
ax1.set_xlabel('Count')
ax1.plot(a,'r')

ax1.set_ylabel('Score')
ax1.bar(range(8),a,color='gray')
ax2.bar(range(8),a,color='gray')
```
